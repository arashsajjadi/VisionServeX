# v3.6 Phase 8: Release Checklist

Final release checklist for VisionServeX v3.6.0.

```python
import visionservex
print('Version:', visionservex.__version__)
assert visionservex.__version__ == '3.6.0', f'Expected 3.6.0, got {visionservex.__version__}'
print('PASS: Version is 3.6.0')
```

```python
# Check pyproject.toml version
try:
    import tomllib
except ImportError:
    import tomli as tomllib
from pathlib import Path

p = Path(visionservex.__file__).parent.parent.parent / 'pyproject.toml'
if p.exists():
    data = tomllib.loads(p.read_text())
    v = data['project']['version']
    print(f'pyproject.toml version: {v}')
    assert v == '3.6.0'
    print('PASS: pyproject.toml version is 3.6.0')
else:
    print('pyproject.toml not found (installed from wheel)')
```

```python
# Check CLI version
import subprocess
result = subprocess.run(['visionservex', '--version'], capture_output=True, text=True)
print('CLI output:', result.stdout.strip())
assert '3.6.0' in result.stdout or '3.6' in result.stdout, f'Version not in output: {result.stdout!r}'
print('PASS: CLI --version shows 3.6.0')
```

```python
# Check locate-anything CLI is registered
result = subprocess.run(['visionservex', 'locate-anything', '--help'], capture_output=True, text=True)
print('Return code:', result.returncode)
assert result.returncode == 0
print('PASS: locate-anything CLI registered')
```
