# v3.6 Phase 1: SAM-ViT-L and SAM-ViT-H ONNX Export (CLI Fix)

Before v3.6, `sam_commands._ONNX_ELIGIBLE` was missing `sam-vit-l` and `sam-vit-h`,
even though both ARE in `onnx_export._SAM_ONNX_ELIGIBLE`. This caused CLI-level ONNX
export to silently fail with 'not ONNX-eligible' for those two variants.

This notebook demonstrates the fixed behavior.

```python
from visionservex.cli.sam_commands import _ONNX_ELIGIBLE, _CHECKPOINT_PATHS

print('ONNX-eligible models:', sorted(_ONNX_ELIGIBLE))
print()
print('Checkpoint paths:')
for mid, path in sorted(_CHECKPOINT_PATHS.items()):
    print(f'  {mid}: {path}')
```

```python
# Verify consistency with onnx_export module
from visionservex.onnx_export import onnx_eligible

eligible = onnx_eligible()
print('onnx_export.onnx_eligible():', eligible)

# All CLI-eligible must be in the export module
for mid in _ONNX_ELIGIBLE:
    status = 'OK' if mid in eligible else 'MISSING'
    print(f'  {mid}: {status}')
```

```python
# Check if sam-vit-l checkpoint is available
from pathlib import Path
for mid, path in _CHECKPOINT_PATHS.items():
    p = Path(path).expanduser()
    exists = p.exists()
    print(f'{mid}: checkpoint {"EXISTS" if exists else "NOT FOUND"} at {p}')
```
