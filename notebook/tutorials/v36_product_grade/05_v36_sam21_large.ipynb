# v3.6 New Execution: SAM2.1-hiera-large

SAM2.1-hiera-large is the largest SAM2.1 variant. It was not individually
executed in v3.5 (only tested as part of the hiera family sweep).
This notebook runs it standalone to produce a new execution record.

```python
from visionservex.vsx import VSX

h = VSX.sam('sam2.1-hiera-large')
print('Model ID:', h.model_id)
print('Status:', h.status())
print('Install extra:', h.explain()['install_extra'])
```

```python
from PIL import Image
import json
from pathlib import Path

img = Image.new('RGB', (256, 256))

try:
    result = h.segment(img, box=[20, 20, 200, 200])
    print('Segmentation result type:', type(result).__name__)
    artifact = {'model_id': 'sam2.1-hiera-large', 'status': 'ok', 'result_type': type(result).__name__}
    Path('/tmp/sam21_large_seg.json').write_text(json.dumps(artifact))
    print('Artifact saved.')
except Exception as e:
    print(f'Execution blocked (expected if SAM2 not installed): {type(e).__name__}: {e}')
```
