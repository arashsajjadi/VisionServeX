# v3.6 New Model: DINO Self-Supervised ViT-S/8 Embedding

The original DINO (Self-Distillation with No Labels) from Meta is distinct from DINOv2.
`facebook/dino-vits8` produces 384-dimensional embeddings (Apache-2.0).

```python
from PIL import Image
import numpy as np

img = Image.new('RGB', (224, 224), color=(100, 150, 200))
print('Image:', img.size)
```

```python
try:
    from transformers import AutoFeatureExtractor, AutoModel
    import torch, time

    model_id = 'facebook/dino-vits8'
    extractor = AutoFeatureExtractor.from_pretrained(model_id)
    model = AutoModel.from_pretrained(model_id)

    inputs = extractor(images=img, return_tensors='pt')
    t0 = time.perf_counter()
    with torch.no_grad():
        outputs = model(**inputs)
    latency_ms = (time.perf_counter() - t0) * 1000

    # CLS token embedding
    embed = outputs.last_hidden_state[:, 0, :].squeeze().numpy()
    print(f'DINO ViT-S/8 embedding shape: {embed.shape}')  # (384,)
    print(f'Latency: {latency_ms:.1f} ms')
    print(f'Embedding norm: {np.linalg.norm(embed):.4f}')

    np.save('/tmp/dino_vits8_embed.npy', embed)
    print('Saved to /tmp/dino_vits8_embed.npy')
except Exception as e:
    print(f'Skipped (missing deps): {e}')
```
