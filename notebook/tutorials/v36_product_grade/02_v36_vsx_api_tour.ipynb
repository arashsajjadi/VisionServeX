# v3.6 VSX Facade: Standardized Python API Tour

VisionServeX v3.6 standardizes the API across all model families.
Every family handle exposes: `explain()`, `status()`, and a family-specific run method.

```python
import visionservex
print('VisionServeX version:', visionservex.__version__)
```

```python
from visionservex.vsx import VSX

# Demonstrate all 5 factory methods
factories = [
    ('sam', 'sam-vit-base'),
    ('dino', 'dinov2-base'),
    ('pipeline', 'grounding-dino-swin-t+sam2-hiera-small'),
    ('cv2', 'opencv-mser-proposals'),
    ('locateanything', 'locate-anything-3b'),
]
for factory, model_id in factories:
    h = getattr(VSX, factory)(model_id)
    print(f'VSX.{factory}({model_id!r}).status() = {h.status()!r}')
```

```python
# Full explain() for SAM
import json
info = VSX.sam('sam2.1-hiera-small').explain()
print(json.dumps(info, indent=2))
```

```python
# Full explain() for LocateAnything (non-commercial)
info = VSX.locateanything('locate-anything-3b').explain()
print('State:', info['state'])
print('Default safe:', info['default_safe'])
print('Commercial safe:', info['commercial_safe'])
print('Warning (first 100 chars):', info['warning'][:100])
```
