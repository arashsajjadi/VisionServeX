# v3.6 New Model: DepthAnything-Small Monocular Depth Estimation

DepthAnything is a new task type for VisionServeX v3.6 — monocular depth estimation.
`depth-anything-small-hf` uses `LiheYoung/depth-anything-small-hf` (Apache-2.0).

```python
from PIL import Image
import numpy as np

img = Image.new('RGB', (640, 480), color=(100, 150, 200))
print('Test image:', img.size)
```

```python
try:
    from transformers import pipeline as hf_pipeline
    import time

    pipe = hf_pipeline('depth-estimation', model='LiheYoung/depth-anything-small-hf')
    t0 = time.perf_counter()
    result = pipe(img)
    latency_ms = (time.perf_counter() - t0) * 1000

    depth = np.array(result['depth'])
    print(f'Depth map shape: {depth.shape}')
    print(f'Depth range: [{depth.min():.2f}, {depth.max():.2f}]')
    print(f'Latency: {latency_ms:.1f} ms')

    np.save('/tmp/depth_anything_small.npy', depth)
    print('Saved depth map to /tmp/depth_anything_small.npy')
except Exception as e:
    print(f'Skipped (missing deps): {e}')
```
