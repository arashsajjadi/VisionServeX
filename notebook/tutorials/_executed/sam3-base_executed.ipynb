**sam3-base** uses the Meta custom SAM License and is **HF-gated**. VisionServeX never mirrors gated weights and never logs your token — BYOT (bring-your-own-token).

## Fresh pip-install verification (imports from site-packages, not local src)

```python
import sys, subprocess, importlib.metadata
VERSION = "3.1.0"
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "-U",
    f"visionservex[classic-ml,cv2-pro]=={VERSION}"])
import visionservex
print("visionservex", importlib.metadata.version("visionservex"))
print("file:", visionservex.__file__)
assert "site-packages" in visionservex.__file__, visionservex.__file__
assert "PycharmProjects/VisionServeX/src" not in visionservex.__file__
```

```python
from visionservex import VSX, VSXError
sam = VSX.sam("sam3-base")
import json; print(json.dumps(sam.explain(), indent=2, default=str))
```

```python
# 1) request HF gated access on the model page  2) export HF_TOKEN=...  3) run.
# Graceful failure without a token (no secret printed):
try:
    sam.segment("image.jpg", box=[0,0,10,10])
except VSXError as e:
    print("state:", e.state); print("next:", e.next_command)
```
