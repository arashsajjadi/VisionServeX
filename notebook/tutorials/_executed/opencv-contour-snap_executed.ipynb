# CV2-Pro tutorial — opencv-contour-snap

Weight-free OpenCV tool **opencv-contour-snap** (Apache-2.0, CPU). Runs on the base install. Use for region proposals / mask refinement in annotation pipelines.

## Fresh pip-install verification (imports from site-packages, not local src)

```python
import sys, subprocess, importlib.metadata
VERSION = "3.1.0"
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "-U",
    f"visionservex[classic-ml,cv2-pro]=={VERSION}"])
import visionservex
print("visionservex", importlib.metadata.version("visionservex"))
print("file:", visionservex.__file__)
assert "site-packages" in visionservex.__file__, visionservex.__file__
assert "PycharmProjects/VisionServeX/src" not in visionservex.__file__
```

```python
import numpy as np
try:
    import cv2
except Exception:
    cv2=None
from visionservex.cv2_pro import run_tool, tool_available, TOOL_LICENSE
print("tool:", "opencv-contour-snap", "| available:", tool_available("opencv-contour-snap"), "| license:", TOOL_LICENSE().get("opencv-contour-snap"))
```

```python
if cv2 is not None and tool_available("opencv-contour-snap")[0]:
    img=(np.random.default_rng(0).normal(120,30,(192,192,3))).clip(0,255).astype("uint8")
    params={"box":[30,30,160,160]} if "opencv-contour-snap"=="opencv-grabcut-plus" else {}
    r=run_tool("opencv-contour-snap", img, **params)
    print({k:v for k,v in r.items() if k!="polygons"})
else:
    print("install: pip install 'visionservex[cv2-pro]'")
```
