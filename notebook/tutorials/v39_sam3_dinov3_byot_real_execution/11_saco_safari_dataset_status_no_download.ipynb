# v3.9 — SACo-Gold, SACo-Silver, SA-FARI dataset status

These are datasets, NOT runtime models. We check access only — no download.

```python
import sys; sys.path.insert(0, 'src')
from visionservex.hf_auth import _detect
tok_raw, _ = _detect()
from huggingface_hub import HfApi
api = HfApi()
for repo in ['facebook/SACo-Gold', 'facebook/SACo-Silver', 'facebook/SA-FARI']:
    try:
        api.auth_check(repo, token=tok_raw)
        print(f'{repo}: access_granted')
    except Exception as e:
        print(f'{repo}: {type(e).__name__} — {str(e)[:60]}')
```

```python
print('SACo and SA-FARI are datasets/evaluation resources, not segmentation model weights.')
print('VisionServeX does NOT download them by default.')
print('No dataset counted as a model execution in the ledger.')
```
