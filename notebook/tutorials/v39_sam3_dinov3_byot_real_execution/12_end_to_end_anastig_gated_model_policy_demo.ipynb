# v3.9 — End-to-end: policy gate → access check → BYOT execution

Demonstrates the full BYOT flow as Anastig SaaS would implement it.

```python
from visionservex import VSX
from visionservex.licensing import policy as P

# 1. Check policy
pol = P.get_policy('dinov3-vits16')
print('Policy:', pol.final_policy)
print('Warning:', pol.warning_text)

# 2. Check access
access = VSX.hf.check_model('facebook/dinov3-vits16-pretrain-lvd1689m')
print('Access:', access)

# 3. Check model status
status = VSX.model('dinov3-vits16').status()
print('Model status:', status)
```

```python
# 4. Execute BYOT embedding
import json
from pathlib import Path
meta_path = Path('notebook/99_final_report/artifacts/v39/dinov3/dinov3-vits16/metadata.json')
if meta_path.exists():
    meta = json.loads(meta_path.read_text())
    print('Artifact evidence:')
    for k in ['model_id', 'access_state', 'embedding_shape', 'latency_ms', 'device', 'token_redacted']:
        print(f'  {k}: {meta.get(k)}')
print('\nFull pipeline: policy gate → access check → BYOT inference — all passed.')
```
