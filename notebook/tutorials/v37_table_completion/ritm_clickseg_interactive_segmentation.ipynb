```python
from visionservex import VSX
for m in ['ritm','clickseg']:
    print(m, VSX.interactive(m).status(), VSX.interactive(m).explain()['commercial_safe'])
# classic refiner that runs today:
from PIL import Image; import numpy as np
img = Image.fromarray((np.random.rand(80,80,3)*255).astype('uint8'))
print('grabcut:', VSX.interactive('grabcut')(img, positive_points=[(40,40)])['status'])
```
