```python
from visionservex.vsx import VSX

# Check all 11 pipeline states
for pid in [
    'grounding-dino-swin-t+sam-vit-base',
    'grounding-dino-swin-b+sam-vit-base',
    'grounding-dino-swin-t+sam2.1-hiera-small',
    'grounding-dino-1.5+sam3-base',
    'dino-x-api+sam3-base',
]:
    info = VSX.pipeline(pid).explain()
    print(f'{pid}: state={info["state"]}')
```

```python
from visionservex import VisionModel
from PIL import Image

img = Image.open('tests/assets/smoke/coco_person_car.jpg').convert('RGB')

# Stage 1: Detect
detector = VisionModel('grounding-dino-swin-t')
det_result = detector.predict(img, text='person . car')
print(f'Stage 1 - Detections: {type(det_result).__name__}')

# Stage 2: Segment with SAM
sam = VisionModel('sam-vit-base')
seg_result = sam.predict(img, boxes=[[50, 50, 300, 400]])
print(f'Stage 2 - Segments: {type(seg_result).__name__}')
```
