# v3.4 SAM ViT-L/H Checkpoint Guide

Documents the exact commands to pull SAM ViT-L and ViT-H checkpoints
and export them to ONNX once available.

**License:** Apache-2.0 (Meta AI original weights)

````python
from visionservex.onnx_export import list_onnx_eligible_models
import json

status = {r['model_id']: r for r in list_onnx_eligible_models()}
for mid in ('sam-vit-l', 'sam-vit-h'):
    print(f"{mid}: {json.dumps(status[mid], indent=2)}")
````

## Pull Commands

```bash
# Pull sam-vit-l (855 MB, Apache-2.0)
visionservex pull sam-vit-l

# Pull sam-vit-h (2.4 GB, Apache-2.0)
visionservex pull sam-vit-h
```

## Export Commands (after pull)

```bash
visionservex sam export-onnx sam-vit-l --out models/sam-vit-l.onnx
visionservex sam export-onnx sam-vit-h --out models/sam-vit-h.onnx
```

````python
import pathlib
# Verify checkpoint paths
for mid in ('sam-vit-l', 'sam-vit-h'):
    s = status[mid]
    p = pathlib.Path(s['checkpoint_path'])
    print(f"{mid}: exists={p.exists()} path={p}")
````

````python
# Export when checkpoint is present
import pathlib
from visionservex.onnx_export import export_sam_decoder_onnx

out_dir = pathlib.Path('notebook/99_final_report/artifacts/v34')
for mid in ('sam-vit-l', 'sam-vit-h'):
    s = status[mid]
    if s['checkpoint_exists']:
        result = export_sam_decoder_onnx(mid, out_dir / f'{mid.replace("-","_")}_decoder.onnx')
        print(f'EXPORTED {mid}: {result["size_mb"]} MB')
    else:
        print(f'CHECKPOINT REQUIRED for {mid}')
        print(f'  Pull: {s["pull_command"]}')
````
