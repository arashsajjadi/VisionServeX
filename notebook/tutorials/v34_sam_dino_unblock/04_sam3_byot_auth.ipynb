# SAM3 BYOT and HuggingFace Auth Workflow

SAM3 (Segment Anything Model 3) is a gated model on HuggingFace Hub.
This tutorial explains the Bring-Your-Own-Token (BYOT) workflow to access SAM3
through VisionServeX.

**Model:** SAM3  
**Status:** `auth_required` (gated HF repo)  
**License:** Meta AI Research License (restricted)

````python
# Cell 2: Status check — shows auth_required
import visionservex as vsx
from visionservex.sam3 import get_sam3_status

status = get_sam3_status()
print(f'SAM3 status      : {status["status"]}')
print(f'HF repo          : {status.get("hf_repo", "N/A")}')
print(f'Gated            : {status.get("gated", False)}')
print(f'Token present    : {status.get("token_present", False)}')
print(f'Access granted   : {status.get("access_granted", False)}')
if status.get('message'):
    print(f'Message          : {status["message"]}')
````

## Token Auth Steps

To access SAM3 weights, follow these steps:

### Step 1 — Request access on HuggingFace
Visit the SAM3 model page and accept the license:
```
https://huggingface.co/facebook/sam3
```

### Step 2 — Generate a HuggingFace token
1. Go to https://huggingface.co/settings/tokens
2. Create a **read** token (or use an existing one).
3. Copy the token.

### Step 3 — Provide the token to VisionServeX
```python
import visionservex as vsx
# Option A: environment variable
import os; os.environ['HF_TOKEN'] = 'hf_your_token_here'

# Option B: huggingface_hub login
from huggingface_hub import login
login(token='hf_your_token_here')

# Option C: vsx.set_token
vsx.set_token('hf_your_token_here')
```

### Step 4 — Pull model
```bash
visionservex pull sam3
```

## What SAM3 Access Grants

Once authenticated and granted access you can use:

| Feature | SAM / SAM2 | SAM3 |
|---|---|---|
| Point prompts | yes | yes |
| Box prompts | yes | yes |
| Mask prompts | limited | yes |
| Video tracking | SAM2 only | yes (improved) |
| Text prompts | no | yes (CLIP-based) |
| 3D segmentation | no | experimental |
| ONNX export | yes | planned |

SAM3 also features:
- Improved IoU estimation
- Multi-mask output with ranked confidence
- Faster hiera backbone variants

````python
# Cell 5: SAM3 capabilities (read from registry)
from visionservex.sam3 import list_sam3_capabilities

caps = list_sam3_capabilities()
print('SAM3 capability registry:')
for cap in caps:
    avail = 'available' if cap['available'] else f'requires: {cap.get("requires", "auth")}'
    print(f'  {cap["name"]:30s} {avail}')
````
