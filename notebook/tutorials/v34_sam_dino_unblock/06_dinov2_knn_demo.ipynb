```python
import json, pathlib
# Load evidence from sprint execution
results = json.loads(pathlib.Path('notebook/99_final_report/artifacts/v34/dinov2_knn_results.json').read_text())
print(json.dumps(results, indent=2))
```

```python
from visionservex import VisionModel
from PIL import Image
import numpy as np

img = Image.open('tests/assets/smoke/coco_person_car.jpg').convert('RGB')

for mid in ['dinov2-small', 'dinov2-base']:
    model = VisionModel(mid)
    result = model.predict(img)
    print(f'{mid}: embed shape={getattr(result, "shape", "check result type")} status=ok')
```

```python
# kNN self-similarity demo
np.random.seed(42)
# Load saved embeddings
for mid, fname in [('dinov2-small', 'dinov2_dinov2_small_embed.npy'), ('dinov2-base', 'dinov2_dinov2_base_embed.npy')]:
    p = pathlib.Path('notebook/99_final_report/artifacts/v34') / fname
    if p.exists():
        emb = np.load(p)
        sim = float(np.dot(emb.ravel(), emb.ravel()) / (np.linalg.norm(emb)**2))
        print(f'{mid}: dim={emb.shape[-1]} knn_self_sim={sim:.3f}')
```
