## What SAM 2.1 Video Tracking Does

SAM 2.1 extends the promptable image segmentation of SAM to videos.
It uses a memory encoder and memory attention mechanism to propagate
object masks across frames:

1. **Frame 0 prompt**: user provides a point, box, or mask on the first frame.
2. **Memory encoding**: SAM 2.1 encodes the prompted frame into object memory.
3. **Propagation**: the model tracks the object forward (and optionally backward)
   through subsequent frames using stored memory tokens.
4. **Output**: per-frame binary masks + IoU confidence scores.

**Supported models:** `sam2.1-hiera-tiny`, `sam2.1-hiera-small`, `sam2.1-hiera-base-plus`, `sam2.1-hiera-large`
