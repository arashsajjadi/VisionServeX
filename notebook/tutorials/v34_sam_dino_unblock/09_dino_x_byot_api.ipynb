# v3.4 DINO-X API BYOT

DINO-X is an external API (no downloadable weights). Requires DEEPDATASPACE_API_KEY.

**License:** External API — proprietary (IDEACVR)

````python
from visionservex.vsx import VSX
import os

info = VSX.dino('dino-x-api').explain()
print(f'state: {info["state"]}')
print(f'auth_required: {info["auth_required"]}')
has_key = bool(os.environ.get('DEEPDATASPACE_API_KEY'))
print(f'DEEPDATASPACE_API_KEY set: {has_key}')
````

## API Setup

```bash
# 1. Get API key from https://deepdataspace.com
export DEEPDATASPACE_API_KEY=your_key_here

# 2. Detect
visionservex dino api dino-x-api path/to/image.jpg --text 'person . car'

# For GroundingDINO 1.5/1.6:
visionservex dino api grounding-dino-1.5 image.jpg --text 'cat'
visionservex dino api grounding-dino-1.6 image.jpg --text 'cat'
```

````python
# Confirm API-only models never pretend to be runnable
for mid in ['dino-x-api', 'grounding-dino-1.5', 'grounding-dino-1.6']:
    info = VSX.dino(mid).explain()
    assert info['state'] in ('external_api_only', 'auth_required')
    assert info['auth_required'] is True
print('PASS: all API-only models correctly show auth_required or external_api_only')
````

````python
# VSXError raised when no key is set
from visionservex.vsx import VSXError
if not os.environ.get('DEEPDATASPACE_API_KEY'):
    try:
        VSX.dino('dino-x-api').detect('dummy.jpg', text='cat')
    except VSXError as e:
        print(f'PASS: VSXError raised state={e.state}')
````
