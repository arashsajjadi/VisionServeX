```python
from visionservex import VisionModel
from PIL import Image

img = Image.open('tests/assets/smoke/coco_person_car.jpg').convert('RGB')

for mid in ['grounding-dino-swin-t', 'grounding-dino-original-swin-t']:
    model = VisionModel(mid)
    result = model.predict(img, text='person . car . vehicle')
    print(f'{mid}: result type={type(result).__name__} status=ok')
```

```python
# Status check for gated/API variants
from visionservex.vsx import VSX
for mid in ['grounding-dino-1.5', 'grounding-dino-1.6', 'dino-x-api']:
    info = VSX.dino(mid).explain()
    print(f'{mid}: state={info["state"]} auth_required={info["auth_required"]}')
```
