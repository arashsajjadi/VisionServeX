## What is LocateAnything-3B?

LocateAnything-3B is a 3-billion-parameter grounded object detection and localization model from NVIDIA (NVlabs). It supports open-vocabulary text-prompted detection and grounding.

VisionServeX integrates LocateAnything-3B as a **restricted sidecar** — the model is NOT included in the default-safe commercial core. It requires:
1. Manual sidecar install (NVlabs/Eagle)
2. `--accept-noncommercial` flag on every run
3. User acknowledgment of the NVIDIA non-commercial license

```python
# Check if sidecar is installed
try:
    import eagle
    print("Eagle sidecar is installed.")
except ImportError:
    print("Eagle sidecar NOT installed. Run:")
    print("  git clone https://github.com/NVlabs/Eagle.git eagle")
    print("  cd eagle/Embodied && pip install -e .")
```

```python
from visionservex.vsx import VSX

# Create a handle — this does NOT load any weights
h = VSX.locateanything('locate-anything-3b')

print('Status:', h.status())
print('License:', h.explain()['license'])
print('Default safe:', h.explain()['default_safe'])
print('Commercial safe:', h.explain()['commercial_safe'])
```
