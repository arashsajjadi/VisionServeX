## HuggingFace Cache (BYOT)

Weights are stored in `~/.cache/visionservex/locate_anything/` (user-local only).
VisionServeX never downloads weights automatically — you must pre-pull them:

```python
from huggingface_hub import snapshot_download
snapshot_download(
    'nvidia/locate-anything-3b',
    cache_dir='~/.cache/visionservex/locate_anything'
)
```
