This notebook shows how to run LocateAnything-3B inference with the Python API.
**Prerequisites**: Eagle sidecar installed (see `02_locate_anything_sidecar_install.ipynb`).

```python
from PIL import Image
import numpy as np
from visionservex.vsx import VSX

# Create a test image (replace with your real image path)
img = Image.new('RGB', (640, 480), color=(100, 150, 200))
print('Image size:', img.size)
```

```python
# Create handle — does NOT load weights
h = VSX.locateanything('locate-anything-3b')
print('Family:', h.family)
print('Status:', h.status())
```

```python
# Run inference — requires accept_noncommercial=True
# This will fail if the Eagle sidecar is not installed.
try:
    result = h.locate(
        img,
        text='cat',
        accept_noncommercial=True  # Required: acknowledge NVIDIA non-commercial license
    )
    print('Result:', result)
except RuntimeError as e:
    print(f'Sidecar not installed: {e}')
except Exception as e:
    print(f'Error: {type(e).__name__}: {e}')
```

```python
# Attempt WITHOUT accept_noncommercial — this should raise VSXError
from visionservex.vsx import VSXError
try:
    h.locate(img, text='cat', accept_noncommercial=False)
    print('ERROR: Should have raised VSXError')
except VSXError as e:
    print('Correctly raised VSXError:', str(e)[:100])
    print('State:', e.state)
    print('Next command:', e.next_command)
```

```python
# CLI equivalent
# visionservex locate-anything run locate-anything-3b image.jpg --text 'cat' --accept-noncommercial --out result.json
import subprocess
result = subprocess.run(
    ['visionservex', 'locate-anything', 'run', 'locate-anything-3b',
     'test.jpg', '--text', 'cat', '--accept-noncommercial', '--out', '/tmp/la_result.json'],
    capture_output=True, text=True
)
print('Return code:', result.returncode)
print('Stderr (first 300 chars):', result.stderr[:300])
```
