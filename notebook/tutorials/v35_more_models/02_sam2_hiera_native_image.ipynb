# v3.5 SAM2-Hiera Native Image Segmentation

Runs all four SAM2-hiera variants (tiny/small/base-plus/large) via VisionModel.

**New in v3.5:** SAM2 (v2) was blocked in v3.4 (`module_not_installed`). Now works via HF transformers Sam2Model.

```python
import pathlib, json, warnings
warnings.filterwarnings('ignore')

artifact = pathlib.Path('notebook/99_final_report/artifacts/v35/sam2_hiera_segmentation.json')
if artifact.exists():
    result = json.loads(artifact.read_text())
    print(json.dumps(result, indent=2))
else:
    print('Artifact not found — run live demo below')
```

```python
from PIL import Image
from visionservex import VisionModel
import time

img = Image.open('tests/assets/smoke/coco_person_car.jpg').convert('RGB')

for mid in ['sam2-hiera-tiny', 'sam2-hiera-small', 'sam2-hiera-base-plus', 'sam2-hiera-large']:
    t0 = time.perf_counter()
    m = VisionModel(mid)
    result = m.predict(img, boxes=[[50, 50, 350, 350]])
    dt = (time.perf_counter() - t0) * 1000
    segs = getattr(result, 'segments', None) or getattr(result, 'masks', None)
    print(f'{mid}: n_segs={len(segs) if segs else 0} dt={dt:.0f}ms')
```
